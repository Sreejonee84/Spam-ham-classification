```python
data.loc[:,'label'] = data.label.map({'ham':0, 'spam':1})
print(data.shape)
data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer()
```

```python
lower_case_documents = []
lower_case_documents = [d.lower() for d in data.sms]
print(lower_case_documents)
```

```python
sans_punctuation_documents = []
import string

for i in lower_case_documents:
    sans_punctuation_documents.append(i.translate(str.maketrans("","", string.punctuation)))
    
sans_punctuation_documents
```

```python
preprocessed_documents = [[w for w in d.split()] for d in sans_punctuation_documents]
preprocessed_documents
```

```python
frequency_list = []
import pprint
from collections import Counter

frequency_list = [Counter(d) for d in preprocessed_documents]
pprint.pprint(frequency_list)
```

```python
doc_array=np.asarray(frequency_list.key())
doc_array
```

```python
doc_array = count_vector.transform(preprocessed_documents).toarray()
doc_array
```

```python
documents = []
documents = [d for d in data.sms]
print(documents)
```

```python
count_vector.fit(documents)
count_vector.get_feature_names()
```
