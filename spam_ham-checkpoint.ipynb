```python
data['tokens'] = data["sms"].apply(lambda x: nltk.word_tokenize(x))
data['pos_tags'] = data['tokens'].apply(lambda x:[t for w, t in nltk.pos_tag(x)])
data['pos_tag_sentence'] = data['pos_tags'].apply(lambda x: ' '.join(x))
```
